```python
# z score is upper and lower values of mean
# comulative probability is
# confidence
# margin
# ppf
import numpy as np
from scipy import stats
data=[10,20,30,40,50,60,70]
mean=np.mean(data)
std=np.std(data)
n=len(data)
confidence = 0.95
z = stats.norm.ppf((1 + confidence) /2)
margin = z * (std / (n ** 0.5))
lower = mean - margin
upper = mean + margin

print("Mean:", mean)
print("95% Confidence Interval:", (lower, upper))
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split , cross_val_predict , KFold
from sklearn.linear_model import LinearRegression , LogisticRegression
from sklearn.metrics import mean_squared_error, r2_score , accuracy_score , confusion_matrix , classification_report
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
X_new = 2 * np.random.rand(100, 1)
y_new = 2 + 3 * X + np.random.randn(100, 1)
```

```python
w , b = 0.0 , 0.0
lr = 0.01
epochs = 1000
n = len(X_new)

for epoch in range(epochs):
  y_pred = w * X_new + b
  dw = (-2/n) * np.sum(X_new * (y_new - y_pred))
  db = (-2/n) * np.sum(y_new - y_pred)
  w = w - lr * dw
  b = b - lr * db

print("weight" , w)
print("bias" , b)
```

```python
plt.scatter(X_new , y_new)
plt.plot(X_new , w * X_new + b , color = "purple")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression")
plt.show()
```

ridge and lasso regression

```python
from sklearn.linear_model import Ridge , Lasso
ridge_model = Ridge(alpha = 1.0)
ridge_model.fit(X_new , y_new)

lasso_model = Lasso(alpha = 1.0)
lasso_model.fit(X_new , y_new)

print("ridge coef" , ridge_model.coef_)
print("lasso coef" , lasso_model.coef_)
```
